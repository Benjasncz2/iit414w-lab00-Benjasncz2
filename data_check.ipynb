## FastF1 Session Load



```python
session = fastf1.get_session(2024, 'Bahrain', 'R')

session.load()

print("\n--- SESSION INFORMATION ---")
print(f"Event:  {session.event['EventName']}")
print(f"Session:  {session.name}\n")

print("--- RESULTS SHAPE ---")
print(f"Number of rows and columns (session.results.shape): {session.results.shape}\n")

print("--- COLUMN NAMES ---")
print(f"Columns in the Results DataFrame:\n{list(session.results.columns)}\n")

print("--- FIRST 5 LAP RECORDS ---")
display(session.laps.head())
```

## Jolpica API Query

```python
url = "https://api.jolpi.ca/ergast/f1/2024/driverstandings/"
response = requests.get(url)
print("--- CONNECTION STATUS ---")
print(f"HTTP Status Code: {response.status_code}")

if response.status_code == 200:
    data = response.json()
    standings_list = data["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
    print("\n--- RESULTS OBTAINED ---")
    print(f"Number of records (drivers) returned: {len(standings_list)}")
    
    print("\n--- TOP 3 DRIVERS ---")
    for i in range(3):
        driver_data = standings_list[i]
        name = f"{driver_data['Driver']['givenName']} {driver_data['Driver']['familyName']}"
        nationality = driver_data['Driver']['nationality']
        points = driver_data['points']
        print(f"{i+1}. {name} | {nationality} | {points} points")
else:
    print("\n[ERROR] Could not connect to the Jolpica API.")
```

## Data Shape Summary

* **(a) Session Choice:** I loaded the **2024 Bahrain Grand Prix (Race)**. I chose this specific session because it was the opening race of the 2024 season, which serves as a great comparative baseline for the car's initial pace before any major mid-season aerodynamic upgrades were introduced by the teams.

* **(b) Dataset Shapes:**
  * **FastF1 `session.results`:** 20 rows × 21 columns (List of 20 drivers and their finishing status).
  * **Jolpica API `driverstandings`:** 23 records (Representing the 23 drivers who participated or scored points across the 2024 season).

* **(c) Informative Observation:** It was highly informative to notice that the FastF1 `session.results` dataframe includes a tremendous amount of non-timing metadata (like `BroadcastName`, `HeadshotUrl`, and `TeamColor`). This means FastF1 is not only useful for pure telemetry training but it also comes pre-packaged with all the categorical labels necessary to build frontend visualizations or dashboards without needing an external database.
